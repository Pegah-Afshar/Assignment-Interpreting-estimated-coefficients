```python
Y=weather_df['apparenttemperature']-weather_df['temperature']
X=weather_df[['humidity','windspeed']]
```

```python
X=sm.add_constant(X)
result=sm.OLS(Y,X).fit()
result.summary()
```

```python
new_df=houseprices_df[['mszoning','street','landcontour']]
```

```python
new_df=pd.get_dummies(new_df,drop_first=True)
```

```python
new_numeric=houseprices_df[['overallqual','grlivarea','garagecars']]
```

```python
X=pd.concat([new_numeric,new_df],axis=1)
Y=houseprices_df['saleprice']
```

```python
X=sm.add_constant(X)
res=sm.OLS(Y,X).fit()
res.summary()
```

mszoning_RH ,street_Pave are not statistically significant. 

```python
#Now, exclude the insignificant features from your model. Did anything change?
new_df=houseprices_df[['landcontour']]
new_df=pd.get_dummies(new_df,drop_first=True)
new_numeric=houseprices_df[['overallqual','grlivarea','garagecars']]
```

```python
X=pd.concat([new_numeric,new_df],axis=1)
Y=houseprices_df['saleprice']
```

```python
X=sm.add_constant(X)
res=sm.OLS(Y,X).fit()
res.summary()
```
